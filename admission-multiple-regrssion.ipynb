```python
# Seperate X Feature Data
```

```python
X = data_df.drop(['Admission Chance'], axis=1)
```

```python
y = data_df['Admission Chance']
```

```python
X.shape
```

```python
y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
# Create train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state = 5)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Train Model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept=True)
```

```python
reg.fit(X_train, y_train)
```

### Model Eval

```python
y_pred = reg.predict(X_test)
```

```python
y_pred.shape
```

```python
print('Linear Model Coefficient (m)', reg.coef_)
print('Linear Model Coefficient (b)', reg.intercept_)
```

```python
plt.scatter(y_test, y_pred, color = 'r')
plt.ylabel('Model Prediction')
plt.xlabel('True values')
plt.title(' Multiple Linear Regression Predictions')
```

```python
# Get number of independent params
k = X_test.shape[1]
# Get length of test data
n = len(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_pred)), '.3f'))
MSE = float(format(mean_squared_error(y_test, y_pred), '.3f'))
MAE = float(format(mean_absolute_error(y_test, y_pred), '.3f'))
r2 = float(format(r2_score(y_test, y_pred), '.3f'))
adj_r2 = float(format(1-(1-r2)*(n-1)/(n-k-1), '.3f'))
MAPE = float(format(np.mean(np.abs((y_test-y_pred)/y_test))*100, '.3f'))
```

```python
print('RMSE: ', RMSE)
print('MSE: ', MSE)
print('MAE: ', MAE)
print('r2: ', r2)
print('adj_r2: ', adj_r2)
print('MAPE: ', MAPE)
```

```python
# Print accuracy
reg.score(X_test, y_test)
```
